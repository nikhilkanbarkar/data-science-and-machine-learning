## matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for ploting in jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig,ax = plt.subplots(figsize = (10,10)) #with hight

# 3. plot the data
ax.plot(x, y)

# 4. customize plot
ax.set(title = "Simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 5. Savw and show(you save the whole figure)
fig.savefig("sample-plot.png")
```

## Making figures with Numpy array

We want:
* Line plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplot

```python
import numpy as np
```

```python
# Create som data
x = np.linspace (0,10,100)
x[:10]
```

```python
# plot the data and create a lineplot
fig , ax = plt.subplots()
ax.plot(x,x**2)
```

```python
# Use some data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another one scatter
fig , ax = plt.subplots()
ax.scatter(x , np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {
    "Almond butter": 10,
    "Penut butter":8,
    "Cashew butter": 12
}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nicks Nut butter shop",
       ylabel="Prices ($)"
      );
```
