## Dot product Example (nut butter sales)

<b>when it is works:</b><br>
    suppose you have a butter shop and here u have various butter drinks... each has different prices. then to calculate the total weekly sales and the prices between them the matrix multiplication is use(Dot product) is used.where we can calculate the total daily sales of various butter drinks . and multiply those with the prices respectively.

```python
nut_file = pd.read_csv("nut-butter-sales.csv")
nut_file
```

```python
np.random.seed(0)
#Number of jars sold
sales_amounts = np.random.randint(20,size=(5,3))
sales_amounts
```

```python
# Create weekly_sales DataFrame
weekly_sales = pd.DataFrame(sales_amounts,
                           index= ["Mon","Tues","Wed","Thurs","Fri"],
                           columns= ["Almond butter","Peanut butter","cashew butter"])
weekly_sales
```

```python
# Creste Prices array
prices = np.array([10, 8, 12])
prices
```

```python
prices.shape
```

```python
# Create butter_prices`DataFrame
butter_prices = pd.DataFrame(prices.reshape(1,3),
                            index = ["prices"],
                            columns = ["Almond butter","Peanut butter","cashew butter"])
butter_prices
```

```python
sales_amounts.shape,prices.shape
```

```python
total_sales = prices.dot(sales_amounts)
```

```python
# shapes are aligned, let's transpose
total_sales = prices.dot(sales_amounts.T)
total_sales
```

```python
# Create Daily_sales
butter_prices.shape, weekly_sales.shape
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
# Doesn't worked becaus of shape, Try transposing

weekly_sales["Total_sales($)"] = daily_sales
weekly_sales
```

```python
weekly_sales["Total_sales($)"] = daily_sales.T
weekly_sales
```
