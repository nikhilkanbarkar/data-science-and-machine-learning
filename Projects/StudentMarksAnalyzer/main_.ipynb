```python
# Calculating the total marks 
subjects =[
    "Maths",
    "Science",
    "Social_sci",
    "English",
    "Marathi",
    "Kannada"
]

dataframe["Total_marks"] = dataframe[subjects].sum(axis=1)
dataframe
```

```python
# Calculating the % of the students
dataframe["Percentage"] = (dataframe["Total_marks"]/600)*100
dataframe
```

Now lets calculate weather student is pass or fail.<br>
<b>conditions</b><br>
student should get above 35 marks in all the subject. then student is pass, otherwise he fails.<br>



```python
def check_result(row):
    failed_subjects=[]

    for subject in subjects:

        if row[subject] < 35:
            failed_subjects.append(subject)
    if len(failed_subjects) == 0:
        return pd.Series(["Pass","None"])

    else:
        return pd.Series([
            "Fail",
            ", ".join(failed_subjects)
        ])
dataframe[["Result", "Failed_subjects"]] = dataframe.apply(
        check_result,
        axis = 1
    )

dataframe
```

<h3>Grade calculation</h3>

<h4>Important Condition</h4>
<p>A student should get a grade only if they pass all subjects.<br>

<b>Rules</b><br>
If the student fails in any one subject → Grade = F<br>
<b>Otherwise:</b><br>
Percentage ≥ 90 → A+<br>
Percentage ≥ 80 → A<br>
Percentage ≥ 70 → B<br>
Percentage ≥ 60 → C<br>
Percentage ≥ 50 → D<br>
Percentage ≥ 35 → E
</p>

```python
def check_grade(row):
    if row["Result"] == "Fail":
        return "F"
    else:
        if row["Percentage"] >= 90:
            return "A+"
        elif row["Percentage"] >= 80:
            return "A"
        elif row["Percentage"] >= 70:
            return "B+"
        elif row["Percentage"] >= 60:
            return "B"
        elif row["Percentage"] >= 50:
            return "C+"
        else:
            return "C"

dataframe["Grade"] = dataframe.apply(
    check_grade,
    axis = 1
)

dataframe
```

<b>Toppers of the class</b>

```python
def toppers_of_class():
    passed_students = dataframe[
        dataframe["Result"] == "Pass"
    ]
    top_10 = passed_students.sort_values(
        by = "Total_marks",
        ascending = False
    ).head(10)
    return top_10

top_10 = toppers_of_class()
top_10
```

<h3>Subject Analysis</h3><br>


```python
def subject_analysis(dataframe, subject):

    total_students = len(dataframe)

    topper = dataframe.loc[
        dataframe[subject].idxmax()
    ]

    passed = (
        dataframe[subject] >= 35
    ).sum()

    failed = (
        dataframe[subject] < 35
    ).sum()

    pass_percentage = (
        passed / total_students
    ) * 100

    fail_percentage = (
        failed / total_students
    ) * 100

    average_marks = dataframe[
        subject
    ].mean()

    highest_marks = dataframe[
        subject
    ].max()

    lowest_marks = dataframe[
        subject
    ].min()

    if pass_percentage >= 85:
        status = "Good"

    elif pass_percentage >= 70:
        status = "Average"

    else:
        status = "Poor"
    return {
    "Subject": subject,
    "Topper": topper["Names"],
    "Topper Marks": highest_marks,
    "Passed": passed,
    "Failed": failed,
    "Pass %": pass_percentage,
    "Fail %": fail_percentage,
    "Average": average_marks,
    "Status": status
    }

analysis = []

for subject in subjects:
    analysis.append(
        subject_analysis(dataframe, subject)
    )

report_df = pd.DataFrame(analysis, index = [1,2,3,4,5,6])

report_df
```
