## Data cleaning and future engineering

```python
sales_df.head(10)
```

```python
# few columns are optional so we can clean it
sales_df = sales_df.drop("Row ID",axis=1)
sales_df = sales_df.drop("Postal Code",axis=1)
```

```python
sales_df = sales_df.drop("Country",axis=1)
```

```python
sales_df
```

```python
sales_df["Order Date"] = pd.to_datetime(sales_df["Order Date"])
sales_df["Ship Date"] = pd.to_datetime(sales_df["Ship Date"])

# Shipping days
sales_df["Shipping Days"] = (sales_df["Ship Date"] - sales_df["Order Date"])
```

```python
# Order Month and Year for calculating sales 
sales_df["Month"] = sales_df["Order Date"].dt.month_name()
sales_df["Year"] = sales_df["Order Date"].dt.year
sales_df["Weekday"] = sales_df["Order Date"].dt.day_name()
```

```python
sales_df
```

In the Superstore dataset:<br>
<br>

Sales = Total amount after discount<br>
Quantity = Number of units sold<br>
Discount = Discount rate (e.g., 0.20 = 20%)<br>
Profit = Profit earned on the sale<br>

```python
sales_df["Selling Price"] = sales_df["Sales"] / sales_df["Quantity"]

sales_df["Original Price"] = (
    sales_df["Sales"] / (1 - sales_df["Discount"])
) / sales_df["Quantity"]

sales_df["Discount Amount"] = (
    sales_df["Original Price"] * sales_df["Quantity"]
) - sales_df["Sales"]
```

```python
sales_df.columns
```

```python
# lets Re-arrange the columns in structured way.

sales_df = sales_df[
    [
        # Order Information
        "Order ID",
        "Order Date",
        "Ship Date",
        "Shipping Days",
        "Ship Mode",

        # Customer Information
        "Customer ID",
        "Customer Name",
        "Segment",

        # Location Information
        "City",
        "State",
        "Region",

        # Product Information
        "Product ID",
        "Category",
        "Sub-Category",
        "Product Name",

        # Sales Information
        "Quantity",
        "Original Price",
        "Discount",
        "Discount Amount",
        "Selling Price",
        "Sales",
        "Profit",

        # Time Features
        "Month",
        "Year",
        "Weekday"
    ]
]
```
