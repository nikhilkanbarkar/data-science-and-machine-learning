## Data cleaning and future engineering

```python
sales_df.head(10)
```

```python
# few columns are optional so we can clean it
sales_df = sales_df.drop("Row ID",axis=1)
sales_df = sales_df.drop("Postal Code",axis=1)
```

```python
sales_df = sales_df.drop("Country",axis=1)
```

```python
sales_df
```

```python
sales_df["Order Date"] = pd.to_datetime(sales_df["Order Date"])
sales_df["Ship Date"] = pd.to_datetime(sales_df["Ship Date"])

# Shipping days
sales_df["Shipping Days"] = (sales_df["Ship Date"] - sales_df["Order Date"])
```

```python
# Order Month and Year for calculating sales 
sales_df["Month"] = sales_df["Order Date"].dt.month_name()
sales_df["Year"] = sales_df["Order Date"].dt.year
sales_df["Weekday"] = sales_df["Order Date"].dt.day_name()
```

```python
sales_df
```

In the Superstore dataset:<br>
<br>

Sales = Total amount after discount<br>
Quantity = Number of units sold<br>
Discount = Discount rate (e.g., 0.20 = 20%)<br>
Profit = Profit earned on the sale<br>

```python
sales_df["Selling Price"] = sales_df["Sales"] / sales_df["Quantity"]

sales_df["Original Price"] = (
    sales_df["Sales"] / (1 - sales_df["Discount"])
) / sales_df["Quantity"]

sales_df["Discount Amount"] = (
    sales_df["Original Price"] * sales_df["Quantity"]
) - sales_df["Sales"]
```

```python
sales_df.columns
```

```python
# lets Re-arrange the columns in structured way.

sales_df = sales_df[
    [
        # Order Information
        "Order ID",
        "Order Date",
        "Ship Date",
        "Shipping Days",
        "Ship Mode",

        # Customer Information
        "Customer ID",
        "Customer Name",
        "Segment",

        # Location Information
        "City",
        "State",
        "Region",

        # Product Information
        "Product ID",
        "Category",
        "Sub-Category",
        "Product Name",

        # Sales Information
        "Quantity",
        "Original Price",
        "Discount",
        "Discount Amount",
        "Selling Price",
        "Sales",
        "Profit",

        # Time Features
        "Month",
        "Year",
        "Weekday"
    ]
]
```

## <b>ANALYSIS</b>

### Overall Sales

```python
total_sales = sales_df["Sales"].sum()
total_sales
```

```python
total_profit = sales_df["Profit"].sum()
total_profit
```

<b>Total Orders</b><br>

Count the total number of unique orders placed.</br>

Since one order can contain multiple products, use unique Order IDs.</br>

```python
total_orders = sales_df["Order ID"].nunique()
total_orders
```

```python
total_quantity_sold = sales_df["Quantity"].sum()
total_quantity_sold
```

```python
average_sales = sales_df["Sales"].mean()
average_sales
```

```python
average_profit = sales_df["Profit"].mean()
average_profit
```

```python
print("=" * 45)
print("        OVERALL SALES ANALYSIS")
print("=" * 45)

print(f"Total Sales         :${total_sales:,.2f}")
print(f"Total Profit        : ${total_profit:,.2f}")
print(f"Total Orders        : {total_orders:,}")
print(f"Total Quantity Sold : {total_quantity_sold:,}")
print(f"Average Sales       : ${average_sales:,.2f}")
print(f"Average Profit      : ${average_profit:,.2f}")

print("=" * 45)
```

### Product Analysis

```python
sales_df
```

<b> Best selling by Quantity

```python
best_selling = (
    sales_df.groupby("Product Name")["Quantity"]
      .sum()
      .sort_values(ascending=False)
)

print(best_selling.head(10))
```

```python
lowest_selling = (
    sales_df.groupby("Product Name")["Quantity"]
            .sum()
            .sort_values()
)
lowest_selling.head(20)
```

<b>most profitable & least profitable </b>

```python
most_profitable = (sales_df.groupby("Product Name")["Profit"]
    .sum()
    .sort_values(ascending = False)
)
most_profitable
```

```python
least_profitable = (
    sales_df.groupby("Product Name")["Profit"]
            .sum()
            .sort_values()
)
least_profitable.head(20)
```

<b>"Top 10 Selling Products By sales

```python
top_product = (sales_df.groupby("Product Name")["Sales"]
                           .sum()
                           .sort_values(ascending=False)
                  )
top_product
```
