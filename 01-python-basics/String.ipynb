### string_creation.py

```python
"""
Creating strings using different quotes.
"""

name = "Python"

course = 'Data Science'

message = """
Python is easy to learn.
It is widely used in AI and Machine Learning.
"""

print(name)
print(course)
print(message)
```

### string_indexing.py

```python
"""
Access characters using indexes.
"""

text = "Python"

print("First Character:", text[0])
print("Second Character:", text[1])
print("Last Character:", text[-1])
print("Second Last Character:", text[-2])
```

### string_slicing.py

```python
"""
Extract portions of a string.
"""

text = "Python Programming"

print(text[0:6])
print(text[7:18])

print(text[:6])

print(text[7:])

print(text[::2])

print(text[::-1])
```

### string_length.py

```python
"""
Find the length of a string.
"""

text = "Machine Learning"

length = len(text)

print("Length:", length)
```

### string_methods.py

```python
"""
Common string methods.
"""

text = "python programming"

print(text.upper())

print(text.lower())

print(text.title())

print(text.capitalize())

print(text.swapcase())
```

### string_search.py

```python
"""
Search text inside a string.
"""

text = "Python Programming"

print("Python" in text)

print("Java" in text)

print(text.find("Programming"))

print(text.find("Java"))
```

### string_replace.py

```python
"""
Replace text inside a string.
"""

text = "I love Java"

updated_text = text.replace("Java", "Python")

print(updated_text)
```

### string_reverse.py

```python
"""
Reverse a string using slicing.
"""

text = input("Enter a string: ")

reversed_text = text[::-1]

print("Reversed String:", reversed_text)
```

### palindrome_check.py

```python
"""
Check whether a string is palindrome.
"""

text = input("Enter a string: ")

cleaned_text = text.lower()

if cleaned_text == cleaned_text[::-1]:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

### vowel_counter.py

```python
"""
Count vowels in a string.
"""

text = input("Enter a string: ")

count = 0

for character in text.lower():

    if character in "aeiou":
        count += 1

print("Number of Vowels:", count)
```

### word_counter.py

```python
"""
Count words in a sentence.
"""

sentence = input("Enter a sentence: ")

words = sentence.split()

print("Total Words:", len(words))
```

### character_frequency.py

```python
"""
Count frequency of each character.
"""

text = input("Enter a string: ")

for character in set(text):

    frequency = text.count(character)

    print(f"{character} : {frequency}")
```

### remove_spaces.py

```python
"""
Remove all spaces from a string.
"""

text = input("Enter a sentence: ")

result = text.replace(" ", "")

print(result)
```

### count_upper_lower.py

```python
"""
Count uppercase and lowercase letters.
"""

text = input("Enter a string: ")

upper_count = 0
lower_count = 0

for character in text:

    if character.isupper():
        upper_count += 1

    elif character.islower():
        lower_count += 1

print("Uppercase Letters:", upper_count)
print("Lowercase Letters:", lower_count)
```
