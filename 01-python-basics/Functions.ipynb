## Functions

### simple_function.py

```python
"""
Basic function without parameters and return value.
"""

def greet():
    print("Welcome to Python Functions")


greet()
```

### function_with_parameters.py`

```python
"""
Function that accepts parameters.
"""

def greet(name):
    print(f"Hello, {name}")


greet("Nikhil")
greet("Rahul")
```

### function_with_return.py

```python
"""
Function that returns a value.
"""

def add(num1, num2):
    return num1 + num2


result = add(10, 20)

print("Sum =", result)
```

### default_arguments.py

```python
"""
Using default argument values.
"""

def student_details(name, course="Python"):
    print("Name:", name)
    print("Course:", course)


student_details("Nikhil")

student_details("Rahul", "Data Science")
```

### keyword_arguments.py

```python
"""
Passing arguments using parameter names.
"""

def employee(name, salary, department):
    print("Name:", name)
    print("Salary:", salary)
    print("Department:", department)


employee(
    department="AI",
    name="Nikhil",
    salary=50000
)
```

### variable_length_arguments.py

```python
"""
Accept multiple values using *args.
"""

def calculate_sum(*numbers):

    total = 0

    for number in numbers:
        total += number

    return total


print(calculate_sum(10, 20))
print(calculate_sum(10, 20, 30))
print(calculate_sum(10, 20, 30, 40))
```

### recursive_factorial.py

```python
"""
Factorial using recursion.
"""

def factorial(number):

    if number == 0 or number == 1:
        return 1

    return number * factorial(number - 1)


num = int(input("Enter a number: "))

print("Factorial =", factorial(num))
```

### lambda_function.py

```python
"""
Anonymous function using lambda.
"""

square = lambda number: number ** 2

print(square(5))
print(square(10))
```

### local_global_scope.py

```python
"""
Difference between local and global variables.
"""

college = "MIT"


def display():

    department = "Computer Science"

    print("College:", college)
    print("Department:", department)


display()

print("College:", college)
```
