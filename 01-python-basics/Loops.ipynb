## Loops

### for_loop_basics.py

```python
"""
Print numbers from 1 to 10 using a for loop.
"""

for number in range(1, 11):
    print(number)
```

### while_loop_basics.py

```python
"""
Print numbers from 1 to 10 using a while loop.
"""

count = 1

while count <= 10:
    print(count)
    count += 1
```

### multiplication_table.py

```python
"""
Generate multiplication table for a given number.
"""

number = int(input("Enter a number: "))

for i in range(1, 11):
    print(f"{number} x {i} = {number * i}")
```

### sum_of_n_numbers.py

```python
"""
Calculate the sum of first n natural numbers.
"""

n = int(input("Enter a number: "))

total = 0

for number in range(1, n + 1):
    total += number

print("Sum =", total)
```

### factorial.py

```python
"""
Calculate factorial of a number.
"""

number = int(input("Enter a number: "))

factorial = 1

for i in range(1, number + 1):
    factorial *= i

print("Factorial =", factorial)
```

### break_statement.py

```python
"""
Stop loop execution when a condition is met.
"""

for number in range(1, 11):

    if number == 6:
        break

    print(number)
```

### continue_statement.py

```python
"""
Skip a particular iteration.
"""

for number in range(1, 11):

    if number == 5:
        continue

    print(number)
```

### pattern_printing.py

```python
"""
Print a simple star pattern.
"""

rows = 5

for i in range(1, rows + 1):

    for j in range(i):
        print("*", end=" ")

    print()
```
