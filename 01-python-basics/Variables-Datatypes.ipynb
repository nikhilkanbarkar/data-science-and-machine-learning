# Python basics

## Variables

### variable_creation.py

```python
"""
Creating and using variables.
A variable stores data in memory.
"""

name = "Nikhil"
age = 21
cgpa = 8.8

print("Name:", name)
print("Age:", age)
print("CGPA:", cgpa)
```

### multiple_assignment.py

```python
"""
Assign multiple values in a single line.
"""

x, y, z = 10, 20, 30

print("x =", x)
print("y =", y)
print("z =", z)
```

### swapping_variables.py

```python
"""
Swap two variables without using a third variable.
"""

a = 10
b = 20

print("Before Swapping")
print("a =", a)
print("b =", b)

a, b = b, a

print("\nAfter Swapping")
print("a =", a)
print("b =", b)
```

### variable_scope.py

```python
"""
Difference between global and local variables.
"""

college = "MIT"

def display():
    department = "Computer Science"

    print("College:", college)
    print("Department:", department)

display()

print("College:", college)
```

## Datatypes

### numeric_types.py

```python
"""
Integer, Float and Complex numbers.
"""

integer_value = 100
float_value = 98.75
complex_value = 2 + 3j

print(integer_value)
print(float_value)
print(complex_value)

print(type(integer_value))
print(type(float_value))
print(type(complex_value))
```

### string_type.py

```python
"""
String datatype examples.
"""

language = "Python"

print(language)

print(language[0])
print(language[-1])

print(language.upper())
print(language.lower())
```

### boolean_type.py

```python
"""
Boolean values.
"""

is_logged_in = True
is_admin = False

print(is_logged_in)
print(is_admin)

print(type(is_logged_in))
```

### type_conversion.py

```python
"""
Convert one datatype into another.
"""

num = "100"

print(type(num))

num = int(num)

print(type(num))
print(num + 50)

percentage = float("89.5")

print(percentage)
```

### type_checking.py

```python
"""
Check datatype using type().
"""

name = "Python"
marks = 90
cgpa = 8.7

print(type(name))
print(type(marks))
print(type(cgpa))
```
